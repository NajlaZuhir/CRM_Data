```python
from faker import Faker
import pandas as pd
import random
```

```python
jobs = ['Engineer', 'Teacher', 'Doctor', 'Business Owner', 'Student', 'Nurse', "Other"]
cities_in_qatar = ["Doha","Al Wakrah","Al Khor","Umm Salal Mohammed","Al Rayyan","Al Daayen","Madinat ash Shamal"]
list_cuisines = ["Italian", "Japanese", "Mexican", "Indian", "French", "Chinese", "Thai", "Arabic", "Other"]

attractions_in_qatar = ["The Pearl-Qatar","Souq Waqif","Museum of Islamic Art","Katara Cultural Village","Aspire Park",
    "Al Zubarah Fort and Archaeological Site","Doha Corniche","Sheikh Faisal Bin Qassim Al Thani Museum",
    "Katara Beach","Qatar National Museum", "Other"]

event_types = ["Festival","Concert","Exhibition","Football event","Championship event","Doha Expo","Workshop",
    "Doha Tour","Marathon", "Other"]

fake_hobbies = ["Reading","Traveling","Photography","Painting","Cooking","Gardening","Playing musical instruments",
    "Hiking","Swimming","Cycling"]
```

### Dataframe

We have added some new columms

```python
fake = Faker()
start_id = 1001
reservation_id = 1

data = []

for i in range(5000):
    record = {
        # identification
        "Tourist_ID": start_id,
        "Reservation_ID": reservation_id,
        "passport_number" : fake.random_int(min=100000000, max=999999999),

        # personal info
        "Name": fake.name(),
        "Age": fake.random_int(18, 80),
        "Gender": fake.random_element(elements=('Male', 'Female')),
        "phone_number": fake.phone_number(),
        "Nationality": fake.country(),
        'Nationality_group': random.choice(['GCC', 'Other Arab', 'Other Africa', 'Other Asia inc. Oceania', 'Europe', 'Americans']),
        'family_size': fake.random_int(min=1, max=8),
        'health_insurance': random.choice(['Yes', 'No']),

        # Date
        'Date_Of_Reservation' : fake.date_between(start_date='today', end_date='-5y'),
        'Date_Of_arival' : fake.date_between(start_date='today', end_date='-5y'),
        'arrival_month': fake.month(),

        # Address info
        'country_address': fake.country(),
        'street_address': fake.street_address(),
        'zipcode': fake.zipcode(),
        'state': fake.state(),
        'city': fake.city(),

        # Occupation and financial info
        "Occupation": fake.random_element(elements=jobs),
        'annual_income': fake.random_int(min=20000, max=150000),

        # personal preferences
        "Hobby": fake.random_element(elements=fake_hobbies),
        'payment_preferences': random.choice(['Credit Card', 'PayPal', 'Bank Transfer']),
        'climate_preferences': random.choice(['Tropical', 'Temperate', 'Arctic']),
        "Best_Attraction": fake.random_element(elements=attractions_in_qatar),
        "Cuisine": fake.random_element(elements=list_cuisines),

        # Travel and Transportation info
        "Preferred_Travel_Seasons": fake.random_element(elements=('Spring', 'Summer', 'Autumn', 'Winter')),
        "Modes_of_Travel": fake.random_element(elements=('Sea', 'Land', 'Air')),
        "Frequency_of_Travel": fake.random_element(elements=('Frequent', 'Occasional', 'Rare')),
        "Transportation_Mode": fake.random_element(elements=('Car', 'Bus', 'Metro')),

        # Visit info
        "Num_of_Visits": fake.random_int(min=1, max=7),
        "Days_of_Stay": fake.random_int(5, 60),
        "Random_Event_Type": fake.random_element(elements=event_types),

        # Accommodation info
        "Room_Preferences": fake.random_element(elements=('Single', 'Double', 'Suite')),
        "Hotel_Type": fake.random_element(elements=('Luxury', 'Mid-Range', 'Budget')),
        "Location": fake.random_element(elements=cities_in_qatar),

        # Engagement and feedback
        "Subscribed_to_Loyalty_Program": fake.boolean(),
        "Rating_of_Services": fake.random_int(min=1, max=5),
        "Total_Spendings": fake.random_int(min=2000, max=30000),

        # Additional fields (agency concern related)
        "customer_tenure": (datetime.now() - pd.to_datetime(record['Date_Of_Reservation'])).days,
        "package_types": fake.random_element(elements=['Sightseeing Tours', 'Adventure Tours', 'Cultural Tours', 'Beach and Resort Tours', 'Luxury and VIP Tours']),
        "package_cost": None,  # Placeholder for now
        "package_price": None,  # Placeholder for now
        "profit": None,  # Placeholder for now
    }

    # Set package_cost and package_price based on the selected package_types
    if record["package_types"] == 'Sightseeing Tours':
        record["package_cost"] = 300
        record["package_price"] = 600
    elif record["package_types"] == 'Adventure Tours':
        record["package_cost"] = 500
        record["package_price"] = 1200
    elif record["package_types"] == 'Cultural Tours':
        record["package_cost"] = 400
        record["package_price"] = 800
    elif record["package_types"] == 'Beach and Resort Tours':
        record["package_cost"] = 800
        record["package_price"] = 1500
    elif record["package_types"] == 'Luxury and VIP Tours':
        record["package_cost"] = 1500
        record["package_price"] = 3000

    # Calculate profit (package_price - package_cost)
    record["profit"] = record["package_price"] - record["package_cost"]

    data.append(record)

    reservation_id += 1
    start_id +=

# Create DataFrame
df = pd.DataFrame(data)

# Save to CSV file
df.to_csv("generated_data.csv", index=False)

# Display DataFrame
print(df.head())
```
